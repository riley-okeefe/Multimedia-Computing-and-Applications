```python
import csv
from matplotlib import image as img
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import numpy as np
import math
from google.colab import drive

drive.mount("/content/drive")
path = "drive/MyDrive/"

# Reading in image file
image = img.imread(path+"outdoor-cat.png")

# Displaying image file
plt.imshow(image,vmin=0, vmax=255)
plt.show()

# Displaying image shape (width x height x # of color components)
print("Image shape: ", image.shape)

# Extracting RGB color components to an array
arr = np.array(image)
arr.resize((arr.shape[0] * arr.shape[1], arr.shape[2]))

# Displaying all colors array shape
print("all colours shape: ", arr.shape)
print("all colours: \n")
print(arr)

# Displaying set of unique colors in image
uniqueColors, indices, counts = np.unique(arr, return_index=True,return_counts=True, axis=0)
print("\nunique colours shape: ", uniqueColors.shape)
print("Unique colours: \n", uniqueColors)

# Get frequency of the unique colors
print("Frequencies: \n", counts)

# Display histogram of unique colors
plt.plot(counts)
plt.show()

# Display the color map of image
arr1 = np.array(uniqueColors)
arr1.resize((275, 275, 3))
plt.imshow(arr1,vmin=0,vmax=255)
plt.show()

# export all the data
all_colours= open(path+"all_colours.csv", "w")
unique_colours = open(path+"unique_colours.csv", "w")
unique_freq = open(path+"unique_freq.csv", "w")

write_all_colours = csv.writer(all_colours)
write_all_colours.writerows(arr)

write_uniq_colours = csv.writer(unique_colours)
write_uniq_colours.writerows(uniqueColors)

write_uniq_freq = csv.writer(unique_freq)
write_uniq_freq.writerows([[count] for count in counts])

# Reduce the color depth to 3-bit for each RGB component
convertedImage = image * 7.0
reducedImage = convertedImage.astype(int)
normalizeImage = reducedImage / 7.0

# Display the image file
plt.imshow(normalizeImage,vmin=0, vmax=255)
plt.show()

# Extracting RGB color components to an array
arr_reduced = np.array(normalizeImage)
arr_reduced.resize((arr_reduced.shape[0] * arr_reduced.shape[1], arr_reduced.shape[2]))

print("all colours shape: ", arr_reduced.shape)
print("all colours: \n")
print(arr_reduced)

# Displaying set of unique colors in reduced image
uniqueColors_reduced, indices_reduced, counts_reduced = np.unique(arr_reduced, return_index=True,return_counts=True, axis=0)
print("\nunique colours shape: ", uniqueColors_reduced.shape)
print("Unique colours: \n", uniqueColors_reduced)

# Get frequency of the unique colors
print("Frequencies: \n", counts_reduced)

# Display histogram of unique colors
plt.plot(counts_reduced)
plt.show()

# Display the colourmap of the reduced image
arr1 = np.array(uniqueColors_reduced)
arr1.resize((8, 8, 3))
plt.imshow(arr1,vmin=0,vmax=255)
plt.show()

# Export all the data
all_colours_reduced= open(path+"all_colours_reduced.csv", "w")
unique_colours_reduced = open(path+"unique_colours_reduced.csv", "w")
unique_freq_reduced = open(path+"unique_freq_reduced.csv", "w")

write_all_colours_reduced = csv.writer(all_colours_reduced)
write_all_colours_reduced.writerows(arr_reduced)

write_uniq_colours_reduced = csv.writer(unique_colours_reduced)
write_uniq_colours_reduced.writerows(uniqueColors_reduced)

write_uniq_freq_reduced = csv.writer(unique_freq_reduced)
write_uniq_freq_reduced.writerows([[count_reduced] for count_reduced in counts_reduced])
```

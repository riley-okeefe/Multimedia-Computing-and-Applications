```python
from IPython.core.interactiveshell import IPython
import pretty_midi
import fluidsynth
from IPython.display import Audio, display
import numpy as np
from google.colab import drive
drive.mount('/content/drive')

path = "drive/MyDrive/"
from scipy.io import wavfile

# Reading MIDI file
filename = "WatermelonMan.mid"

pm = pretty_midi.PrettyMIDI(path+filename)

# Convert the MIDI file to audio data and save as wav file
sr = 44100
wave = pm.fluidsynth(fs=sr)
Audio(wave, rate=sr)
wavfile.write(path+"audio_master.wav", sr, wave)

# Seperate tracks into individual objects
print(pm.instruments)
print(len(pm.instruments))

midi_objects = []

# Seperate tracks/instruments into individual MIDI objects
for instrument in pm.instruments:
  if instrument.is_drum:
    pm_obj = pretty_midi.PrettyMIDI()
    pm_obj.instruments.append(instrument)
  if not instrument.is_drum: #if the instrument is not a drum raise each note by an octave
    pm_obj = pretty_midi.PrettyMIDI()
    inst = pretty_midi.Instrument(program = instrument.program, is_drum = False, name = instrument.name)
    for note in instrument.notes:
       new_note = pretty_midi.Note(velocity = note.velocity, pitch = note.pitch + 12, start = note.start, end = note.end)
       inst.notes.append(new_note)
    pm_obj.instruments.append(inst)

  midi_objects.append(pm_obj)



print(len(midi_objects))
print(midi_objects[0].instruments)

# Converting each MIDI object to audio data
audio_list = []
for i,item in enumerate(midi_objects):
  wav_obj = midi_objects[i].fluidsynth(fs=sr)
  audio_list.append(wav_obj)

print(len(audio_list))

# Writing each audio data to wav file
# audio 1
wavfile.write(path+midi_objects[0].instruments[0].name+".wav", sr, audio_list[0])
# audio 2
wavfile.write(path+midi_objects[1].instruments[0].name+".wav", sr, audio_list[1])
# audio 3
wavfile.write(path+midi_objects[2].instruments[0].name+".wav", sr, audio_list[2])
# audio 4
wavfile.write(path+midi_objects[3].instruments[0].name+".wav", sr, audio_list[3])
# audio 5
wavfile.write(path+midi_objects[4].instruments[0].name+".wav", sr, audio_list[4])
# audio 6
wavfile.write(path+midi_objects[5].instruments[0].name+".wav", sr, audio_list[5])
# audio 7
wavfile.write(path+midi_objects[6].instruments[0].name+".wav", sr, audio_list[6])
# audio 8
wavfile.write(path+midi_objects[7].instruments[0].name+".wav", sr, audio_list[7])
# audio 9
wavfile.write(path+midi_objects[8].instruments[0].name+".wav", sr, audio_list[8])
# audio 10
wavfile.write(path+midi_objects[9].instruments[0].name+".wav", sr, audio_list[9])
# audio 11
wavfile.write(path+midi_objects[10].instruments[0].name+".wav", sr, audio_list[10])
# audio 12
wavfile.write(path+midi_objects[11].instruments[0].name+".wav", sr, audio_list[11])
# find min length of audio tracks in audio_list
min_length = min(len(audio) for audio in audio_list)

# shorten audio tracks to have same length
for i,audio in enumerate(audio_list):
  if len(audio) > min_length:
    audio_list[i] = audio[:min_length]
  elif len(audio) < min_length:
    audio_list[i] = np.pad(audio, (0,min_length - len(audio)), 'constant')

# combine all audio data
combined_wave = np.sum(audio_list, axis=0)
# Save the combined audio as "combined.wav"
wavfile.write(path + "combined.wav", sr, combined_wave)
```

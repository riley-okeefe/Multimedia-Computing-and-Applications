```python
import nltk
nltk.download()
from google.colab import drive
drive.mount('/content/drive')

path = "drive/MyDrive/"

filename = "friends_ep624.txt"

with open(path+filename) as f:
  text_data = f.read()
```

```python
import nltk
from nltk import word_tokenize
import numpy as np

w_tokens = np.array(word_tokenize(text_data))
print(w_tokens)

with open(path+"w_tokens.txt", "w", newline="") as token_file:
  for token in w_tokens:
    token_file.write(token + "\n")
```

```python
# removing anything that isn't a word from the list
import re

is_word = []
for token in w_tokens:
  if re.search(r"[a-zA-Z]", token):
    is_word.append(token)

w_tokens = is_word
print(w_tokens)
```

```python
# removing common stopwords
from nltk.corpus import stopwords
nltk.download("stopwords")
stop_words = stopwords.words("english")

# w_tokens_lower = np.char.lower(w_tokens)

not_stopwords = []
# deals with upper/lower case words
for token in w_tokens:
  if not token in stop_words and re.search(r"[^a-z]", token) or not token in stop_words and re.search(r"[^A-Z]", token):
    not_stopwords.append(token)

w_tokens = not_stopwords

print(w_tokens)
```

```python
# removing peoples names and postfixes

# names = ["rachel", "ross", "thompson", "monica", "chandler", "joey", "richard", "phoebe", "elizabeth"]
names = ["Rachel", "Ross", "Thompson", "Monica", "Chandler", "Joey", "Richard", "Phoebe", "Elizabeth"]
no_name = []
for token in w_tokens:
  if not token in names:
    no_name.append(token)
print(no_name)
w_tokens = no_name
```

```python
# removing postfixes and stop words not removed by previous task i.e. 'I'
postfixes = ["n't", "'s", "'re", "'m", "na", "gon", "I",
             "You", "The", "It", "What", "She", "No", "Scene"]
not_postfix = []
for token in w_tokens:
  if not token in postfixes:
    not_postfix.append(token)
print(not_postfix)
w_tokens = not_postfix
```

```python
from ast import Str
from nltk.probability import FreqDist
import matplotlib.pyplot as plt
from collections import Counter

f_distribution = FreqDist(w_tokens)

all_common = f_distribution.most_common(len(w_tokens))

m_common = f_distribution.most_common(20)

all_words, all_freq = zip(*all_common)
words, frequencies = zip(*m_common)

print(all_common)
print(words)
print(frequencies)

plt.bar(words, frequencies)
plt.xticks(rotation=90)
plt.show()

with open(path+"BoW.csv", "w") as file:
  for i, j in all_common:
    file.write(i+","+str(j)+"\n")
print(len(w_tokens))
```

```python
from nltk import sent_tokenize

# tokenize original corpus into sentences
s_tokens = np.array(sent_tokenize(text_data))
print(s_tokens)

with open(path+"s_tokens.txt", "w") as file:
  for token in s_tokens:
    file.write("------\n" + token + "\n")
```

```python
# Extracts the sentences that include the words
# in the 20 most frequently used words, and saves
# these extracted sentences as "sentences_with_top20.txt"
# with dashed lines separating them as above.
from collections import OrderedDict
print(words)
sentences_top_twenty = []
for token in s_tokens:
  for word in words:
    if word in token:
      sentences_top_twenty.append(token)
# remove the duplicates
print(sentences_top_twenty)
result = list(OrderedDict.fromkeys(sentences_top_twenty)) #works to remove duplicates
print(result)
# write to file
with open(path+"sentences_with_top20.txt", "w") as file:
  for token in result:
    file.write("------\n" + token + "\n")
```
